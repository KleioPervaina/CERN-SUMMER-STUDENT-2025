```python
# Path to your bsrt_analysis folder
DATADIR = Path("/eos/project/l/lhc-lumimod/LuminosityFollowUp/2025/bsrt_analysis/10739")

# Parquet files to read
parquet_files = ["B1_injection.parquet", "B2_injection.parquet"]

# Read each into a DataFrame
dfs = {}
for fname in parquet_files:
    path = DATADIR / fname
    if not path.exists():
        raise FileNotFoundError(f"File not found: {path}")
    dfs[fname] = pd.read_parquet(path)

# Inspect contents
for fname, df in dfs.items():
    print(f"\n=== {fname} ===")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print(df.head(), "\n")
```
